### Choosing the target column: 

So we have 4 tests that show weather the person has cancer or not. None of these tests are 100% reliable and often their results don't agree with each other. However if even at least one test shows positive then we cannot ignore it because saying that the person who has cancer that he does not have it is really bad (false negatives are really bad in our case). So I decide to do so that if at least one test shows positive than I say that person is diagnosed with positive (average of four is >0.24) than person has cancer (value 1).  
On the other side it might be better to choose somehting like if at least 2 test show positive then we say that person is diognosed with cancer. However 2 issues: 1) we can ommit some real ill people 2) then I get too few 1's in my data set ( somethign like 7%) and it is very hard to work and make predictions on such a data. should be at least above 10% and my new column has just over 10% of 1's. 



```python
Avg_result=[]
for i in range(data.shape[0]):
    summ=data.Hinselmann[i]+data.Schiller[i]+data.Citology[i]+data.Biopsy[i]
    avg=summ/4
    Avg_result.append(avg)
data['Avg_result']=Avg_result
```

```python
for i in range(data.shape[0]):   
    if Avg_result[i]>0.24:
        Avg_result[i]=1
    else:
        Avg_result[i]=0
data['Avg_result']=Avg_result
```

# Combine smoke columns

I want to combine smokes(years) and Smokes(packcks/year) to one column that would be smokes packs total -> where I multipy Skomes(yaars) by Smokes(packs/years). I do it so than we can see if there is correlation with cancer.

```python
Smoked_total=[]
for i in range(data.shape[0]):
    product=data['Smokes (years)'][i]*data['Smokes (packs/year)'][i]
    Smoked_total.append(product)
data['Smoked_total']=Smoked_total
data['Smoked_total']
```

```python
data.drop(['Smokes (years)','Smokes (packs/year)'],axis=1, inplace=True)
```

I decide to look where the number of unique values is 1 (all values are the same) and drop these columns because of their value does not change these features have no practical meaning. 

```python
cn=[]
for i in range(data.shape[1]):
    col=data.columns[i]
    cn.append(data[col].nunique())
print(data.columns)
print(cn)
```

```python
i=0
for n in data.columns:
    if cn[i]==1:
        print(data.columns[i])
        print(data[data.columns[i]][0])
        data.drop([data.columns[i]],axis=1,inplace=True)
        cn.remove(1)
        i=i
    else:
        i=i+1
```

So we see there were 2 columns with unique values = 1 and I dropped them

Next I want to devide my columns in numerical and cathegorical because I wnat to fill NaNs in columsn but its better to use mode for cathegorical and median for numerical columns.

```python
numerical_columns = []
categorical_columns =[]
i=0
for n in data.columns:
    if cn[i]>2:
        numerical_columns.append(n)
    elif cn[i]==2:
        categorical_columns.append(n)
    i=i+1    
```

```python
data.count(axis=0)
```

So we can see above that there are a lot of missing values. We fill them and then observe that now we have all the values!
As it was advied I fill numerical columns with median and cathegorical columns with mode

```python
for col in numerical_columns:
    data[col].fillna(data[col].median(), inplace=True)
for col in categorical_columns:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score, KFold
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
import statistics as st
```

Here I do my first most simple fit. And my data is not great. So we want to optimize the parameters to get best results for the LinearRegression model. 

```python
model = make_pipeline (StandardScaler(), LogisticRegression())
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print('Accuracy', accuracy_score(y_test, y_predict))
print('precision', precision_score(y_test, y_predict))
print('recall', recall_score(y_test, y_predict))
print('f1', f1_score(y_test, y_predict))
```

I want to look at the values for cross validation. But then I use GridSearchCV to optimize my  parameters
I also use stratifyKfold so that when I split my set into multipe sets I keep the same ration between 1's and 0's in each set! Really important for my imbalanced data 

```python
SefKFold = StratifiedKFold(n_splits=5, random_state=34, shuffle=True)
cv = SefKFold.split(X_train, y_train)
model = make_pipeline (StandardScaler(), LogisticRegression())
scoresAccuracy = cross_val_score(model, X_train, y_train, cv=cv)
print('Accuracy', st.mean(scoresAccuracy))

cv = SefKFold.split(X_train, y_train)
model = make_pipeline (StandardScaler(), LogisticRegression())
scoresF1 = cross_val_score(model, X_train, y_train, scoring='f1', cv=cv)
print('F1', st.mean(scoresF1))

cv = SefKFold.split(X_train, y_train)
model = make_pipeline (StandardScaler(), LogisticRegression())
scoresRecall = cross_val_score(model, X_train, y_train, scoring='recall', cv=cv)
print('Recall', st.mean(scoresRecall))
```

Many classifiers has class_weight parameter. It allows to increase the weight of a smaller class (ref. Notes lecture 9) - Linear regression has it and it really impoved my score

Here I choose best penalty and C for linear regression



```python
SefKFold = StratifiedKFold(n_splits=5, random_state=72, shuffle=False)
cv = SefKFold.split(X_train, y_train)

model = make_pipeline (StandardScaler(), LogisticRegression(class_weight = 'balanced'))
parameters = {'logisticregression__penalty':['l1','l2'],
             'logisticregression__C': np.linspace(0.00000001,100, 11)}
grid = GridSearchCV(model, parameters, cv=cv, n_jobs=-1, verbose=True, scoring='f1')
grid.fit(X_train, y_train)
print(grid.best_score_)
print(grid.best_params_)
```

I get much  much better result with my new parameters on the train set (comparing to the results on the test set above). But lets see for the results with my new parameters on the test set

```python
model = make_pipeline (StandardScaler(), LogisticRegression(penalty='l2', C=10**-8, class_weight = 'balanced'))
model.fit(X_train, y_train)

y_predict = model.predict(X_test)
print('Accuracy', accuracy_score(y_test, y_predict))
print('precision', precision_score(y_test, y_predict))
print('recall', recall_score(y_test, y_predict))
print('f1', f1_score(y_test, y_predict))
```

Aersome. This looks better. We cannot expect high f1 because we do not have much data and it is very imbalanced... But we managed to improve out result. Also I am happy with  my recall

```python
clf = make_pipeline (StandardScaler(), XGBClassifier())
clf.fit(X_train, y_train)
print('Accuracy', accuracy_score(y_test, clf.predict(X_test)))
print('recall', recall_score(y_test, clf.predict(X_test)))
print('f1', f1_score(y_test, clf.predict(X_test)))
```

Lets find best parameters and imporove our results. 

On this website (https://www.analyticsvidhya.com/blog/2016/03/complete-guide-parameter-tuning-xgboost-with-codes-python/) there is shown great tactics on how to choose and fit parameters for XGBoost classifier. (Thanks to TA for the link). So I kind of followed their suggestion plan of attack. 
first I look at n_estimators, max_depth and learning_rate parameters - most common ones. The ranges choses are actually chosen after multiple itterations so now I know that best is around those ranges.

```python
SefKFold = StratifiedKFold(n_splits=4, random_state=72, shuffle=True)
cv = SefKFold.split(X_train, y_train)
parameters = {'xgbclassifier__n_estimators': [400 ,500, 700, 1000, 1200], 
                    'xgbclassifier__max_depth': range(3, 10),
                    'xgbclassifier__learning_rate': [0.01, 0.05],
                    'xgbclassifier__objective': ['binary:logistic'] 
                   }                  
clf = GridSearchCV(make_pipeline(StandardScaler(), XGBClassifier()), param_grid = parameters, scoring='f1', 
                   n_jobs=-1, cv=cv,verbose=True)
clf.fit(X_train, y_train)
print(clf.best_score_)
print(clf.best_params_)
```

That was the first step. Now we look at more parameters becasue I am not happy with the result.. Next step is to look at min_child_weight and gamma parameters with other parameters that we already found

```python
cv = SefKFold.split(X_train, y_train)
parameters = {'xgbclassifier__n_estimators': [1000], 
                    'xgbclassifier__max_depth': [4],
                    'xgbclassifier__learning_rate': [0.05],
                    'xgbclassifier__objective': ['binary:logistic'],
                    'xgbclassifier__min_child_weight': [1, 3, 5],
                    'xgbclassifier__gamma': [0, 0.05]
                   }                  
clf = GridSearchCV(make_pipeline(StandardScaler(), XGBClassifier()), param_grid = parameters, scoring='f1', 
                   n_jobs=-1, cv=cv,verbose=True)
clf.fit(X_train, y_train)
print(clf.best_score_)
print(clf.best_params_)
```

Last step is to look at the regularization parameters alpha and lambda

```python
cv = SefKFold.split(X_train, y_train)
parameters = {'xgbclassifier__n_estimators': [1000], 
              'xgbclassifier__max_depth': [4],
              'xgbclassifier__learning_rate': [0.05],
              'xgbclassifier__objective': ['binary:logistic'],
              'xgbclassifier__min_child_weight': [1],
              'xgbclassifier__gamma': [0],
              'xgbclassifier__reg_alpha': [1e-5, 1e-2, 0.1, 1, 100],
              'xgbclassifier__reg_lambda': [1e-5, 1e-2, 0.1, 1, 100],
                   }         
clfGS = GridSearchCV(make_pipeline(StandardScaler(), XGBClassifier()), param_grid = parameters, scoring='f1', 
                     n_jobs=-1, cv=cv,verbose=True)
clfGS.fit(X_train, y_train)
print(clfGS.best_score_)
print(clfGS.best_params_)
```

We actually impoved our results with new regularization parameters! Great. Now I have set of the parameters that are best for my XGBoost model. Lets see how good this model is on my test set. 

```python
clf = make_pipeline (StandardScaler(), XGBClassifier(n_estimators = 1000, max_depth = 4, learning_rate = 0.05, 
                                                     objective = 'binary:logistic', gamma = 0, min_child_weight = 1,
                                                       reg_lambda = 1, reg_alpha = 0.01))
clf.fit(X_train, y_train)
print('Accuracy', accuracy_score(y_test, clf.predict(X_test)))
print('recall', recall_score(y_test, clf.predict(X_test)))
print('f1', f1_score(y_test, clf.predict(X_test)))
```

Not as good as linear regression unfortunately. But we managed to improve the reslts of my tree model with specific parameters.



Now lets see at some graphics. First I needPlot the dependency of GridSearchCV score on max_depth. To do so I run GridSearchCV with my best parameters I fount and with max_depth parameters in range from 1 to 10. (more than 10 can lead to overfitting so I do not go higher). 

```python
cv = SefKFold.split(X_train, y_train)
parameters = {'xgbclassifier__n_estimators': [1000], 
              'xgbclassifier__max_depth': range(1,11),
              'xgbclassifier__learning_rate': [0.05],
              'xgbclassifier__objective': ['binary:logistic'],
              'xgbclassifier__min_child_weight': [1],
              'xgbclassifier__gamma': [0],
              'xgbclassifier__reg_alpha': [0.01],
              'xgbclassifier__reg_lambda': [1],
                   }         
clfGS = GridSearchCV(make_pipeline(StandardScaler(), XGBClassifier()), param_grid = parameters, scoring='f1', 
                     n_jobs=-1, cv=cv,verbose=True)
clfGS.fit(X_train, y_train);
```

```python
print(clfGS.best_score_)
print(clfGS.best_params_)
```

```python
scores = clfGS.cv_results_['mean_test_score']
max_depth = range(1,11)
plt.plot(max_depth, scores)
```

As expected best is max_depth 4 (as we have seen before). But the behaviour is like interesting. No obvious dependancy. However, we get very low scores for too low and too hight max_depth


Last thing to do is to plot PR curves. I found a great code example in one of the lectures. So I implemented it for my case (nobody said we cannnot use lecture codes)      (Precision-Recall curves should be used when there is a moderate to large class imbalance -my case)

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
y_prob = model.predict_proba(X_test)
y_prob = y_prob[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_prob)
plt.step(recall, precision, color='b', alpha=0.2,
         where='post')

y_pred = model.predict(X_test)
plt.plot(recall_score(y_test, y_pred), 
         precision_score(y_test, y_pred), 
         '*r', ms=10)

plt.fill_between(recall, precision, alpha=0.2, color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
print('Area under the curve: ', metrics.auc(recall, precision))
plt.title('PR curve for linear regression')
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
y_prob = clf.predict_proba(X_test)

y_prob = y_prob[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_prob)
plt.step(recall, precision, color='b', alpha=0.2,
         where='post')

y_pred = clf.predict(X_test)
plt.plot(recall_score(y_test, y_pred), 
         precision_score(y_test, y_pred), 
         '*r', ms=10)

plt.fill_between(recall, precision, alpha=0.2, color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('PR curve for XGBoost classifier')
print('Area under the curve: ', metrics.auc(recall, precision))
```

These plots are pretty informative. We see as expected that at recall 0 precission is 1. And as we get higher recall, precission is going down. Also we look at the areas under the curves becasue they summarize the skill of a model across thresholds. It is like average precission of the model. As we got better results for the linear regreassion model we also get greater area under the curve for linear regression tha for XGBoost model.
