Example of normalization:

```python
from remizova_inclass1 import normalize_image, recognize_digits


for fname in os.listdir('./examples2')[::5]:
    fname = './examples2/' + fname
    image_full = io.imread(fname, as_gray=True, plugin='matplotlib')
    
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image_full)
    
    image = normalize_image(image_full)
    plt.subplot(1, 2, 2)
    plt.imshow(image)
    
    plt.show()
```

In my case, recognize_digits have parameter `complete_solve` -- not only it gets the form of sudoku, it also solves it (or throws the error :)) and gets the image of solved one. I decided to check it on a couple of images and draw result if recognition was perfect:

```python
from tqdm import tqdm_notebook
images = []
all_digits = []
for fname in tqdm_notebook(os.listdir('./examples2'))[:5]:
    fname = './examples2/' + fname
    image_full = io.imread(fname, as_gray=True, plugin='matplotlib')
    
    image = normalize_image(image_full)
    
    try:
        digits = recognize_digits(image, complete_solve=True)
        all_digits.append(digits)
        images.append(image)
    except ValueError:
        continue
```

```python
for i in range(len(images)):
    plt.figure(figsize=(5, 5))
    plt.imshow(images[i])
    plt.show()
    print(all_digits[i][0])
```

Couple of examples of solutions filled:

```python
for i in range(len(images)):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(images[i])
    
    plt.subplot(1, 2, 2)
    plt.imshow(all_digits[i][1])
    plt.show()
```
