#  Normalize the image


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import cv2 as cv
from skimage.transform import resize

%matplotlib inline
```

```python
from skimage import io
from skimage.feature import canny
```

# resizing + edge detection

```python
#fname = './IMG_1286.JPG'
fname = './IMG_1264.JPG'
image_full = io.imread(fname, as_gray=True, plugin='matplotlib')
io.imshow(image_full)
```

```python
from skimage.transform import rescale
image_scaled = rescale(image_full, 1)
edges = canny(image_scaled)
io.imshow(edges)
```

```python
from skimage.morphology import dilation, disk

selem = disk(1)
edges = dilation(edges, selem)
io.imshow(edges)
```

```python
import cv2

edges = (edges).astype(np.uint8)
```

```python
ext_contours = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0]

fig, ax = plt.subplots()
ax.imshow(edges, cmap=plt.cm.gray)

for n, contour in enumerate(ext_contours):
    contour = np.array(contour).squeeze()
    if contour.ndim > 1:
        ax.plot(contour[:, 0], contour[:, 1], linewidth=2)
```

```python
ext_contours = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]

fig, ax = plt.subplots()
ax.imshow(edges, cmap=plt.cm.gray)

for n, contour in enumerate(ext_contours):
    contour = np.array(contour).squeeze() 
    if contour.ndim > 1:
        ax.plot(contour[:, 0], contour[:, 1], linewidth=2)
```

```python
contour = max(ext_contours, key=cv2.contourArea)

print(contour.shape)
contour = contour.squeeze()
print(contour.shape)
```

```python
fig, ax = plt.subplots()
ax.imshow(image_scaled, cmap=plt.cm.gray)
ax.plot(contour[:, 0], contour[:,  1], 'r')
```

# finding the corners

```python
epsilon = 0.05 * cv2.arcLength(contour, True)
corners = cv2.approxPolyDP(contour, epsilon, True).squeeze()
```

```python
fig, ax = plt.subplots()
ax.imshow(image_scaled, cmap=plt.cm.gray)
ax.plot(corners[:, 0], corners[:, 1], '*r')
```

```python
corners
```

# transformation

```python
points_desired = [[image_scaled.shape[1],0],
                          [0,0],
                          [0, image_scaled.shape[0]],
                          [image_scaled.shape[1],image_scaled.shape[0]]]

points_desired = np.array(points_desired)
points_current = corners
```

```python
from skimage.transform import ProjectiveTransform, warp
```

```python
tform = ProjectiveTransform()
tform.estimate(points_desired, points_current)
image_warped = warp(image_scaled, tform)
img = image_warped
img = resize(img,(2394,2394))
io.imshow(img,cmap=cm.gray)
```
