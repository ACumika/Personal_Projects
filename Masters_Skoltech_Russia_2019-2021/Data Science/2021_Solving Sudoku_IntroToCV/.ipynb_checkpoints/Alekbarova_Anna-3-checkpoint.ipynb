# How to find a rectangular object with the highest area

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

%matplotlib inline
```

```python
from skimage import io
from skimage.feature import canny
```

## Step 1: resizing + edge detection

```python
fname = './IMG_1286.JPG'
# fname = './scrable.jpg'
image_full = io.imread(fname, as_gray=True, plugin='matplotlib')
io.imshow(image_full)
```

```python
from skimage.transform import rescale
image_scaled = rescale(image_full, 0.1)
edges = canny(image_scaled)
io.imshow(edges)
```

##  Step 2a: find all contours and selecting the largest one (optional)

#### ontours can be discontinious, in these cases we need to apply a simple morphological operation

```python
from skimage.morphology import dilation, disk

selem = disk(1)
edges = dilation(edges, selem)
io.imshow(edges)
```

##  Step 2b: find all contours and selecting the largest one

```python
import cv2

# we need to convert image to uint to apply findContours
edges = (edges).astype(np.uint8)
```

#### OpenCV provides a function to find all contours ...

```python
ext_contours = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0]

fig, ax = plt.subplots()
ax.imshow(edges, cmap=plt.cm.gray)

for n, contour in enumerate(ext_contours):
    contour = np.array(contour).squeeze()
    if contour.ndim > 1:
        ax.plot(contour[:, 0], contour[:, 1], linewidth=2)
```

#### or just external ones 

```python
ext_contours = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]

fig, ax = plt.subplots()
ax.imshow(edges, cmap=plt.cm.gray)

for n, contour in enumerate(ext_contours):
    contour = np.array(contour).squeeze() # we need to remove one dim, se below
    if contour.ndim > 1:
        ax.plot(contour[:, 0], contour[:, 1], linewidth=2)
```

#### For each contour we can calculate the Area and select the contour with the highest value

```python
contour = max(ext_contours, key=cv2.contourArea)

# we need to remove one unnecessary dimension
print(contour.shape)
contour = contour.squeeze()
print(contour.shape)
```

```python
fig, ax = plt.subplots()
ax.imshow(image_scaled, cmap=plt.cm.gray)
ax.plot(contour[:, 0], contour[:,  1], 'r')
```

## Step 3: finding the corners

#### Another function looks for a similar close curve which should be as simple as possible 

```python
# epsilon allow us to control the max deviation from the original curve 
# here we use 5% of the total curve lenth 
epsilon = 0.05 * cv2.arcLength(contour, True)
corners = cv2.approxPolyDP(contour, epsilon, True).squeeze()
```

```python
fig, ax = plt.subplots()
ax.imshow(image_scaled, cmap=plt.cm.gray)
ax.plot(corners[:, 0], corners[:, 1], '*r')
```

```python
from skimage.transform import ProjectiveTransform, warp
```

```python
tform = ProjectiveTransform()
tform.estimate(points_desired, points_current)
image_warped = warp(image_scaled, tform)
io.imshow(image_warped)
```
